# Create Table

```python
from mage_ai.settings.repo import get_repo_path
from mage_ai.io.bigquery import BigQuery
from mage_ai.io.config import ConfigFileLoader
from os import path

config_path = '/Volumes/dealshare_mvp_new/default/io_config/io_config.yaml'
config_profile = 'default'

query = f"""CREATE OR REPLACE TABLE `dealshare-d82f7.{staging_dataset}.databricks_mvp_{table_name}_staging`
            AS
            SELECT *
            FROM `dealshare-d82f7.{target_dataset}.{table_name}`
            WHERE {modified_date_column} >= '2025-09-01 00:00:00'
            AND {modified_date_column} < '2025-12-01 00:00:00';
            """

BigQuery.with_config(ConfigFileLoader(config_path, config_profile)).execute(query)
```
